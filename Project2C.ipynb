# Scientific Question: What is the common genetic ancestor of enzyme luciferase and green fluorescent protein?
Luciferase and GFP are bioluminecent proteins. 

Luciferase is an oxidative enzyme with light being produced from a series of chemical reactions. The luciferase rxn is Mg+2 dependent and requires substrates luciferin, ATP, and O2. The first reaction is the formation of an enzyme-bound luciferyl-adenylate. During the second reaction, the luciferyl-adenylate undergoes an oxidative decarboxylation which results in the production of CO2, oxyluciferin, AMP, and light. When excess substrates are added to firefly luciferase the reaction produces a flash of light that is proportional to the quantity of luciferase in the reaction mixture.

Green fluorescent protein is a light emmiting protein with an emmision wavelength relative to the excitation wavelength. The barrel structure of the GFP harbors a chromophore molecule responsible for the emmission of a photon of lower energy which is emmitted as a result of a higher energy photon exciting an electron in the chromophore structure and emmitting a photon when the chromophore electron falls down an energy level. 

The pairwise sequence analysis was done on genetic information from luciferase and GFP. These sequences were collected from NCBI database. Another database used in this project is the Protein Data Bank that is used in conjunction with ngl viewer to display protein structures within this jupyter notebook. 

# Scientific Hypothesis: If luciferase and GFP do not have genetic similarities, then the protein structures will not share structural or catalytic domains. 

The sequences were analyzed with a pairwise sequence method. The data used for this analysis was formatted in a fasta file which was downloaded from the NCBI website. From the Protein Data Bank the pdb file was downloaded and visualized using the ngl viewer in the visualization of the data.

# Bioinformatics Analysis


```python
from Bio import pairwise2
from Bio.Seq import Seq
from Bio import SeqIO

GFP = SeqIO.read("GFP.fasta", "fasta")
#print(GFP)
#print(len(GFP_seq))
#Code Check to see if the fasta file read in correctly
GFP_seq = GFP.seq

Luciferase = SeqIO.read("Luciferase.fasta", "fasta")
#print(Luciferase)
#print(len(Luciferase_seq))
#Code Check to see if the fasta file read in correctly
Luciferase_seq = Luciferase.seq

# Finding similarities
alignments = pairwise2.align.globalxx(GFP_seq, Luciferase_seq)
  
# Showing results
for match in alignments:
    print(match)
```

# Plotting the Results
Create a dot plot to show where the 2 sequences line up.

```python
X, Y = GFP, Luciferase
X, Y = np.array(list(X)), np.array(list(Y))

plt.imshow(X==Y[:,None])        # the magic happens here, contourf should work similarly
plt.xticks(np.arange(len(X)), X)
plt.yticks(np.arange(len(Y)), Y)
plt.show(GFP,Luciferase)
```
